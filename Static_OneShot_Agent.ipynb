```python
from smolagents import CodeAgent, OpenAIServerModel
from openai import OpenAI
import os

# Replace 'YOUR_HF_TOKEN_HERE' with your actual Hugging Face API Token
#os.environ['HF_TOKEN'] = '' #set in env config

# Set the NVIDIA API key from an environment variable. Insert key here.
os.environ['NVIDIA_API_NGC_KEY'] = ''

#Check that the API key works by invoking a simple command from the endpoint
client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=os.environ["NVIDIA_API_NGC_KEY"],
)

models = client.models.list()

for m in models.data[:10]:
    print(m.id)
```

```python
# Initialize a model. Taking meta/llama-3.3-70b-instruct

# Create an agent with no tools


from smolagents import CodeAgent, LiteLLMModel

model = LiteLLMModel(
    model_id="nvidia_nim/meta/llama-3.3-70b-instruct",
    api_key=os.environ["NVIDIA_API_NGC_KEY"],
    api_base="https://integrate.api.nvidia.com/v1",
    max_tokens=512,
)

agent = CodeAgent(
    tools=[],
    model=model
)

result = agent.run("Calculate the sum of numbers from 1 to 10")
print(result)
```
